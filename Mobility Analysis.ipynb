```python
def extractUniqueDates(data):
    dates_set = set()
    for row in data:
        datestring = row[1] # Timestamp is at the 2nd position (i.e., index 1)
        # Creating datetime as a dateutil object to pasre date from the ISO 8601 format time string
        datetime = dateutil.parser.parse(datestring) 
        date = datetime.date() # Extracting the date from datetime
        dates_set.add(date) # Adding date to the dates set.
    dates_set = sorted(dates_set)    
    return dates_set

def extract7Dates(data):
    data_week = list()
    for d in data:
        data_week.append(d)
        if(len(data_week) == 7): # If we have 7 days then break the loop
            break
    return data_week
```

##### Calculate trip distances

```python
# Import the Haversine package for computing the distance between two points 
from haversine import haversine
import operator
```

```python
def extractTrips(data, dates):
    day_trips = list()

    for d in dates:
        d_data = list()
        #print("d = ", d)
    
        for row in data:
            datestring = row[1]
            datetime = dateutil.parser.parse(datestring) 
            date = datetime.date()
            
            if (date == d): 
                print("date = ", date)
                d_data.append(row)

        sorted_data = sorted(d_data, key=operator.itemgetter(1), reverse=False)

        # Compute the distance for the trips on day "d".
        d_trips = list()
        
        # Note we have the range from first to second last item as we need to compute the distance with next entry.
        # But for the last entry we will not have any next entry.
        
        for i in range(0, len(sorted_data)-1): 
            d1 = sorted_data[i] # get i-th data instance
            d2 = sorted_data[i+1] # get (i+1)-th data instance
            lat1 = float(d1[2]) # Extract latitude from 2nd index (i.e., 3rd item) and type-cast it to float.
            lon1 = float(d1[3]) # Extract latitude from 3nd index (i.e., 4rd item) and type-cast it to float.
            geo_point1 = [lat1, lon1] # Create a list of lat and lon.
            lat2 = float(d2[2]) 
            lon2 = float(d2[3]) 
            geo_point2 = [lat2, lon2]
            dist = haversine(geo_point1, geo_point2, miles = False)
            d_trips.append(dist) # Append the dist to d_trips list.


        # Add trips on day "d" to day_trips list.
        day_trips.append(d_trips)
    return day_trips
```

#### Calculate maximum daily distance

```python
def maxDistance(data):
    daily_max = list() # a list of maximum distance of daily trips
    
    for trips in data:
        dmax = 0 # initialising the md variable with 0.
        
        for trip in trips:
            if(trip > dmax): # if current trip is greater than md then replace md with current trip.
                dmax = trip
        daily_max.append(dmax)
    return daily_max
```

#### Calculate average daily distance

```python
def averageDistance(daily_trips):
    daily_avg = list()
    
    for trips in daily_trips:
        if(len(trips) == 0):
            avg = 0
        else:
            avg = sum(trips)/len(trips)
        daily_avg.append(avg)
    return daily_avg
```

#### Calculate daily SD

```python
def dailySD(daily_trips):
    daily_sd = list()
    
    for trips in daily_trips:
        if(len(trips) == 0):
            sd = 0
        else:
            ad = sum(trips)/len(trips)
            # compute standard deviation
            sd = 0
            for trip in trips:
                sd = sd + (trip - ad)**2
            sd = (sd/len(trips))**0.5
        daily_sd.append(sd)
    return daily_sd
```

### Create K-Means Class

```python
import random

""" Code courtesy of Mirco Musolesi
    This class can be used for clustering location data by using k-means clustering approach """
class KMeans:

    def __init__(self, k):
        self.k = k          # number of clusters
        self.centroids = None   # means of clusters
        
    """ Return the id of the cluster closest to the input """
    def classify(self, p):
        # taken distance with first centroid as the min_dist (minimum distance) from the given point
        min_dist = self.get_squared_distance(p, self.centroids[0]) 
        cent = 0
        # checking if any other centroid has distance (from the given point) less than min_dist
        for i in range(1, self.k):
            dist = self.get_squared_distance(p, self.centroids[i])
            if min_dist > dist:
                cent = i
        return cent

    
    def train(self, p_list):
        # randomly selecting K points as centroids
        self.centroids = random.sample(p_list, self.k)
        # this is an empty list which will store the cluster ids
        c_ids = []
        #print(self.centroids)
        #count = 0

        while True:
            # Compute new cluster ids
            new_c_ids = []
            for p in p_list:
                new_c_ids.append(self.classify(p))
            
            # If no cluster ids have changed, we're done. No need to compute the cluster ids again.
            if (c_ids == new_c_ids):
                #print(clusters)
                #print(c_ids, new_c_ids)
                return

            # Otherwise store the new cluster ids, update the centroids and compute new cluster ids
            c_ids = new_c_ids

            # updating the centroids (i.e. self.centroids)
            for k in range(self.k):
                # find all points in the cluster k and store them in k_points
                k_points = []
                for i in range(len(c_ids)):
                    if c_ids[i] == k:
                        k_points.append(p_list[i])
                if k_points: # this will avoid divide-by-zero error if k_points is empty
                    # compute centroid of cluster k by using points in k_points
                    self.centroids[k] = self.get_centroid(k_points) 
            #count = count + 1
            
        print((set(c_ids))) 

    """ Computes squared distance between two points """
    def get_squared_distance(self, p1, p2):
        dist_kms = haversine(p1, p2, miles=False) 
        return dist_kms ** 2
    
    """ Computes centroid of the given points """
    def get_centroid(self, all_points):
        lats = list()
        lons = list()
        # get list of all lats and lons
        for p in all_points:
            lats.append(p[0])
            lons.append(p[1])
        # compute the mean of lats and lons
        c_lat = float(sum(lats))/len(lats)
        c_lon = float(sum(lons))/len(lons)
        centroid = (c_lat, c_lon)
        return centroid
```

Implement K-Means clustering on each time interval list

```python
def performKMeans(points_list, K):
    c_points_list = list()
    
    for points in points_list:
        print('start')
        clusters = 0
        
        while clusters != K:
            km = KMeans(K) # everytime create a new class instance to reinstantiate all variables to default
            km.train(points) # train the model

            # now classify the points
            c_ids = list() 

            for p in points:
                c_id = km.classify(p)
                c_ids.append(c_id)

            # Creating a list of K empty lists and storing points of each cluster in separate sub-lists
            k_lists = []
            for k in range(0,K):
                k_list = []
                for i in range(0, len(points)):
                    if c_ids[i] == k:
                        k_list.append(points[i]) # append all points that are in cluster k
                k_lists.append(k_list)
            
            clusters = len(set(c_ids))
            print(clusters)
        # add the clustered points to c_points_list
        c_points_list.append(k_lists)
        
    return c_points_list
```

```python
from __future__ import division
from haversine import haversine
import random

# The k-means algorithm takes a dataset X of N points as input, together with a parameter K specifying how many 
# clusters to create. The output is a set of K cluster centroids and a labeling of X that assigns each of the 
# points in X to a unique cluster. All points within a cluster are closer in distance to their centroid than they 
# sare to any other centroid.

""" It is a class for clustering location data by using k-means clustering approach """
class KMeans:

    
    def __init__(self, k):
        self.k = k          # number of clusters
        self.centroids = None   # means of clusters
        

    """ Return the id of the cluster closest to the input """
    def classify(self, p):
        # taken distance with first centroid as the min_dist (minimum distance) from the given point
        min_dist = self.get_squared_distance(p, self.centroids[0]) 
        cent = 0
        # checking if any other centroid has distance (from the given point) less than min_dist
        for i in range(1, self.k):
            dist = self.get_squared_distance(p, self.centroids[i])
            if min_dist > dist:
                cent = i
        return cent

    
    def train(self, p_list):
        # randomly selecting K points as centroids
        self.centroids = random.sample(p_list, self.k)
        # this is an empty list which will store the cluster ids
        c_ids = []
        
        while True:
            # Compute new cluster ids
            new_c_ids = []
            for p in p_list:
                new_c_ids.append(self.classify(p))
            
            # If no cluster ids have changed, we're done. No need to compute the cluster ids again.
            if c_ids == new_c_ids:                
                return

            # Otherwise store the new cluster ids, update the centroids and compute new cluster ids
            c_ids = new_c_ids

            # updating the centroids (i.e. self.centroids)
            for k in range(self.k):
                # find all points in the cluster k and store them in k_points
                k_points = []
                for i in range(len(c_ids)):
                    if c_ids[i] == k:
                        k_points.append(p_list[i])
                if k_points: # this will avoid divide-by-zero error if k_points is empty
                    # compute centroid of cluster k by using points in k_points
                    self.centroids[k] = self.get_centroid(k_points) 
                
                   

    """ Computes squared distance between two points """
    def get_squared_distance(self, p1, p2):
        dist_kms = haversine(p1, p2, miles=False) 
        return dist_kms ** 2
    
    """ Computes centroid of the given points """
    def get_centroid(self, all_points):
        lats = []
        lons = []
        # get list of all lats and lons
        for p in all_points:
            lats.append(p[0])
            lons.append(p[1])
        # compute the means of lats and lons
        c_lat = float(sum(lats))/len(lats)
        c_lon = float(sum(lons))/len(lons)
        centroid = (c_lat, c_lon)
        return centroid
    
    def get_K(self):
        return self.k    
```
